# Functions in Python

## Introduction to Functions

Definition:

A function is a block of code that performs a specific task. Functions help in organizing code, reusing code, and improving readability.

```python
## parameter function

def greet(name):
    print(f"Hello {name}! Welcome to learning AI")


# Calling function
greet("Puneet")
```

```python
## default function

def greet(name = "guest"):
    print(f"Hello {name}!")

greet()
```

```python
## variable length arguments

# positional args

def info(*args):
    print(args)

lst = [1,2,3,5,6,7]
info(lst)
```
