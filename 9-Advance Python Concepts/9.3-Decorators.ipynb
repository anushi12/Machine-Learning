### Decorators

Decorators are a powerful and flexible feature in Python that allows you to modify the behavior of a function or class method. They are commonly used to add functionality to functions or methods without modifying their actual code. 

```python
## Function Copy
## Closures
## Decorators
```

```python
### Function Copy

def welcome(name):
    return f"Hi!! {name} Welcome to Learning Machine Learning"

greet = welcome     ## function copy
greet("Anu")
```

```python
## Closures

def main_welcome(message):
    # message = "Welcome"
    def sub_welcome():
        print("Welcome to ML")
        print(message)
        print("Hope you are enjoying")
    return sub_welcome
```

```python
greet = main_welcome("Welcome")
greet()
```

### Practice questions Closure

```python
### Write a function counter() that returns a function which, when called repeatedly, prints the count starting from 1 and increments each time.

def counter():
    cnt = 0
    def calls():
        nonlocal cnt
        cnt += 1
        print(cnt)
    return calls

caller = counter()
caller()
caller()
caller()
```

```python
## Create a function make_multiplier(n) that returns a function which multiplies any input by n.

def make_multiplier(n):
    def multiplies(m):
        print(n * m)
    return multiplies

double = make_multiplier(2)
triple = make_multiplier(3)

double(5)
triple(5)
```

#### Decorator


```python
def function(func):
    def sub_function():
        print("before function runs")
        func()
        print("after function runs")
    return sub_function

@function           ## decorator 
def greet():
    print("Hello everyone!")

greet()     ## greet = function(greet)
```
