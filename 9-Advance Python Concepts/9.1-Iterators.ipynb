### Iterators

Iterators are advanced Pyhton concepts that allow for efficient looping and memory management. Iterators provide a way to access elements of a collection sequentially without exposing the underlying structure.

```python
my_list = [1,2,4,5,7,3,9]

for i in my_list:
    print(i)
```

```python
## iterator 

iterator = iter(my_list)
print(type(iterator))
```

```python
next(iterator)
```

```python
iterator = iter(my_list)
```

```python
try:
    print(next(iterator))
except StopIteration:
    print("There are no elemets left in list")
```
