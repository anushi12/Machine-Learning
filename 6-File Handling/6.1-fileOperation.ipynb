## Read a whole file

- The open() function is used for working with files. It takes two paramenters; filename and mode.
- There are four different methods for opening a file: "r" - read, 'a' - append, 'w' - open a file for writing, 'x'- creates the specified file
- Syntax to open a file = open("demo.txt")
- f = open("demo.txt", 'rt') ---> 'r' for read and 't' for text
- read() method is for reading the content of the file.
- with statement --> we don't have to worry about closing the file as with statement takes care of that.

- It is always necessary to close the file when done working with it.
- f.close()
- The reason being changes made to a file may not show until you close the file due to buffering.
- readline() ---> returns the first line.

## Write to an existing file

- To write to an existing file, we must add the parameter 'a' or 'w' with the open() function.
- open("demo.txt", 'w') ---> overwrites the entire file
- f.write("xyzzz") ----> for writing to an exisiting file or new file

- open("demo.txt, 'a') ---> will append to the end of the file

```python
## create a new file 
## "x" is used for creating a new file, it will return an error if the file already exist
f = open("myfile.txt", "x")

```

```python
with open('myfile.txt', 'r') as f:
    content = f.read()
    print(content)
```

```python
## Read a file line by line

with open("myfile.txt", "r") as f:
    for line in f:
        print(line.strip()) ## strip() removes the newline character
```

```python
## Writing in a file

with open("myfile.txt", "w") as f:
    f.write("Wrapping myself inside the duvet,\nand sipping while reading 'The Black Dress',\nevery moment feels undeniably beautiful.")
```

```python
## "a" ---> doesn't overwrites the file instead append it at the end of the content

with open("myfile.txt", "a") as f:
    f.write("\nWrapping myself inside the duvet,\nand sipping while reading 'The Black Dress',\nevery moment feels undeniably beautiful.")
```

```python
### writing a list of lines to a file
lines = ["What a beautiful day it is!\n","Sunday morning's are special at it purest form,\n","Nothing can be better than sitting outdoor and take some time to glance into self."]

with open("myfile.txt",'a') as f:
    f.writelines(lines)
```

```python
## Binary file

file = open("Binary.bin", "x")
```

```python
## writing to a binary file

data = b'\x00\x01\x02\x03\x04'
with open("Binary.bin", 'wb') as file:
    file.write(data)
```

```python
## reading binary file

with open("Binary.bin",'rb') as file:
    content = file.read()
    print(content)
```

```python
## read the content from src text file and write to a dest text file
## copying a text file

with open("myfile.txt", 'r') as src_file:
    content = src_file.read()
with open("destination.txt", 'w') as dest_file:
    dest_file.write(content)
```

```python
# read a text file and count the number of lines, words, and characters.

def count_text_file(file_path):
    with open("myfile.txt", 'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count

file_path = "myfile.txt"
lines, words, chars = count_text_file(file_path)
```

The W+ mode in Python is used to open a file for both reading and writing. If the file does not exist, it will create a new file. Else, it's content will be truncated(i.e., the file will be overwritten).

```python
## Writing and then reading a file

with open("myfile.txt","w+") as file:
    file.write("\nAs the day wraps and the night takes over, it's time to look through and cherish every moment.\n")
    file.write("New day with lots of energy and mind full of fresh thoughts to put into one.\n")

    ## Move the file cursor to the beginning
    file.seek(0)

    ## read the content of the file
    content = file.read()
    print(content)
```
