1. Write a program that takes two numbers from the user and divides them.
Handle ZeroDivisionError and ValueError.

```python
try:
    num1 = int(input("enter first number"))
    num2 = int(input("enter second number"))
    res = num1 / num2
except ZeroDivisionError as ex:
    print(ex)
    print("enter denominator greater than 0.")
except ValueError as ex1:
    print(ex1)
    print("Enter a valid number")

```

2. Create a program that opens a file and reads its content.
Handle:

FileNotFoundError

PermissionError

Any other generic exception

```python
try:
    file = open("myfile.txt",'r')
    content = file.read()
    print(content)
except FileNotFoundError as ex:
    print(ex)
except PermissionError as ex1:
    print(ex1)
    print("Cannot edit! File is in read mode only")
except Exception as ex2:
    print(ex2)
```

4. Demonstrate how a finally block runs even if an exception occurs or not —
Example: divide numbers and print “Program Finished” no matter what.

```python
try:
    a = int(input("enter any value"))
    b = int(input("enter any number"))
    res = a/b
except ZeroDivisionError as ex:
    print(ex)
finally:
    print(res)
    print("Program Finished")
```

5. Write a program that uses try, except, else, and finally together.

else should print “No errors occurred” when everything runs fine.

```python
try:
    a = int(input("enter any value"))
    b = int(input("enter any number"))
    res = a/b
except ZeroDivisionError as ex:
    print(ex)
else:
    print("No errors Occurred")
finally:
    print(res)
    print("Program Finished")
```
