### Data Visualization With Seaborn

Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attactive and informative satistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code.

```python
import seaborn as sns
```

```python
## Basic plotting with Seaborn

tips = sns.load_dataset('tips')
tips
```

```python
## Scatter plot
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x = 'total_bill', y="tip", data=tips)
plt.title("Scatter plot of total bill Vs tips")
```

```python
## line plot

sns.lineplot(x="size",y="total_bill",data=tips)
plt.title("line plot of total bill by size")
```

```python
## Categorical plots

sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Bar plot of total bill by day")
plt.show()
```

```python
## Box plot

sns.boxplot(x="day", y='total_bill', data=tips)
```

```python
## violin plot

sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
## histogram

sns.histplot(tips['total_bill'], bins= 8, kde=True)
```

```python
## KDE plot

sns.kdeplot(tips['total_bill'], fill=True)
```

```python
## pair plot

sns.pairplot(tips)
```

```python
## Heatmap

corr=tips[['total_bill', 'tip', 'size']].corr()
corr
```

```python
sns.heatmap(corr, annot=True) # cmap='coolwarm'
```

```python
## plot total revenue by product

plt.figure(figsize=(8,4))
sns.barplot(x = 'Product', y='Revenue',data=sales_df, estimator=sum)
plt.title('Total sales by product')
plt.xlabel("product")
plt.ylabel("Revenue")
plt.xticks(rotation = 45)
plt.show()
```
