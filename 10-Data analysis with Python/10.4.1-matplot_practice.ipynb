```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("coffee_shop_sales_matplot.csv")
df.head()
```

```python
## Line Plot : Daily Revenue trend

plt.figure(figsize=(8,4))
plt.plot(df["Day"],df["Revenue"],marker = "o")
plt.xlabel("Days")
plt.ylabel("Revenue")
plt.title("Daily revenue Trend")
plt.xticks(rotation = 45)
plt.grid(True)
plt.show()
```

```python
## Bar plot : Quantity Sold by Product

plt.figure(figsize=(8,4))
quantity_sold = df.groupby(["Product"])["Quantity"].sum()
quantity_sold.plot(kind="bar", color="darkorange")
plt.xlabel("Products")
plt.ylabel("Quantity")
plt.title("Quantity sold by product")
plt.show()
```

```python
## total revenue by category

plt.figure(figsize=(8,4))
total_revenue = df.groupby('Category')['Revenue'].sum()
total_revenue.plot(kind = 'bar', color=['#FFB6C1','#87CEEB'])
# plt.bar(total_revenue.index, total_revenue.values, color=['#FFB6C1','#87CEEB'])
plt.title("total revenue by product")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.show()
```

```python
## Average quantity sold by day of week

avg_quantity = df.groupby('Day')['Quantity'].mean()
plt.scatter(avg_quantity.index, avg_quantity.values, color = 'darkgreen')
plt.title("Average quantity sold by Day of week")
plt.xlabel("Days")
plt.ylabel("Avg quantity")
plt.show()
```
