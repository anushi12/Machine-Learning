```python
## if - else statements

age = int(input("enter your age: "))

if(age >= 18):
    print("you are allowed to vote")
else:
    print("you are a minor")
```

```python
## elif statement

age = int(input("enter your age"))

if age < 13:
    print("you are a child")
elif age <18:
    print("you are a teenager")
else:
    print("you are an adult")
```

```python
## nested conditional statements

num = int(input("enter your num: "))

if num > 0:
    print("number is positive")
    if num % 2 == 0:
        print("num is even")
    else: 
        print("num is odd")
else:
    print("number is either 0 or negative")
```

```python
## use cases

'''
1. determine if a year is a leap year
'''

year = int(input("enter year: "))

if year % 4 == 0:
    if year % 100 ==0:
        if year % 400 == 0:
            print(year, "is a leap year")
        else:
            print(year, "is not a leap year")
    else:
        print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```

```python
## simple calculator

num1 = int(input("enter first num: "))
num2 = int(input("enter second num: "))
operation = input("enter operations (+, -, %, *, /): ")

if operation == '+':
    print(num1 + num2)
elif operation == '-':
    print(num1 - num2)
elif operation == '%':
    print(num1 % num2)
elif operation == '*':
    print(num1 * num2)
elif operation == '/':
    if num2 != 0:
        print(num1/num1)
    else:
        print("error! division by zero.")
else:
    print('enter valid operations')
```
