```python
## for loop

for i in range(10):
    print(i)
```

```python
range(1,10,2)  ## start number, end number, step size

for i in range(1, 10, 2):
    print(i)
```

```python
for i in range(10,1, -1):
    print(i)
```

```python
## strings

name = "Puneet"

for i in name:
    print(i)
```

```python
## while loop

count = 0

while count < 5:
    print(count)
    count = count + 1
```

```python
## loop control statements

## break

for i in range(1,10):
    if i == 5:
        break
    else:
        print(i)
```

```python
## continue

for i in range(1,10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
## pass

## pass is a null opeartion; it does nothing

for i in range(5):
    if i == 3:
        pass
    print(i)
```

```python
## Nested loop

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
## calculate sum of all natural numbers

sum = 0
for i in range(11):
    sum = sum + i
print(sum)
```

```python
sum = 0
i = 0
while i <= 10:
    sum = sum + i
    i = i + 1
print(sum)
```

```python
## prime numbers between 1 and 100

for num in range(1,101):
    if num > 1:
        for i in range(2,num):
            if num % i == 0:
                break
        else:
            print(num)
```
