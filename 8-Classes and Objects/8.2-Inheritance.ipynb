Inheritance is a fundamental concept in Object-Oriented Programming (OOP) that allows a class to inherit attributes, and methods from another class.

```python
## inheritance
## parent class

class Car:
    def __init__(self,windows,doors,enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype
    
    def drive(self):
        print(f"the person will drive the {self.enginetype} car")
    
```

```python
car1 = Car(4,4,"petrol")
car1.drive()
```

```python
## child class
## super() ---> allows you to access and reuse the parent class's behavior without repeating the code.
## when a child class overrides a method (like __init__) of its parent, the parent's method doesn't run automatically.
class Tesla(Car):
    def __init__(self,windows,doors,enginetype, is_selfdriving):
        super().__init__(windows,doors,enginetype)  
        self.is_selfdriving = is_selfdriving
    def selfdriving(self):
        print(f"Tesla supports self-driving : {self.is_selfdriving}")
```

```python
tesla1 = Tesla(4,4,"electric","True")
tesla1.selfdriving()
tesla1.drive()
```

```python
## multiple inheritance --- when a class inherits from more than 1 base class

## Base class 1
class Animal:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def speak(self):
        print("base class 1 method called")

# base class 2
class Pet:   
    def __init__(self,owner):
        self.owner = owner

# derived class
class Dog(Animal, Pet):
    def __init__(self,name,age,owner):
        Animal.__init__(self,name,age)
        Pet.__init__(self,owner)
    def speak(self):
        return (f"{self.name} says woof...woof!")

```

```python
dog1 = Dog("tyson",1,"Anushka")
print(f"owner: {dog1.owner}")
dog1.speak()
```
