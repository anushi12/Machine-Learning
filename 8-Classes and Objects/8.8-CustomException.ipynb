```python
class Error(Exception):
    pass

class dobException(Error):
    pass
```

```python
year = int(input("enter your birth year: "))
age = 2025- year

try:
    if age <= 30 and age >= 20:
        print("You can apply for the competitive exams")
    else:
        raise dobException
except dobException:
    print("Age does not fall under the criteria")
```

```python
## handling negative numbers

class NegativeNumber(Exception):
    def __init__(self,value):
        self.value = value
        super().__init__(f"Negative numbers are not allowed: {value}")

    
def sqrt(value):
    if value < 0:
        raise NegativeNumber(value)
    else:
        return value ** 0.5


try:
    number = int(input("enter a number: "))
    result = sqrt(number)
    print(result)
except NegativeNumber as e:
    print(e)
except ValueError:
    print(ValueError)
finally:
    print("program finished")
```

```python
### Create a program that raises a custom exception TooHotError if the temperature entered is above 50°C and TooColdError if it’s below 0°C.

class TooHotError(Exception):
    def __init__(self,temp):
        super().__init__(f"Temperature is extremely above {temp} deg")

class TooColdError(Exception):
    def __init__(self,temp):
        super().__init__(f"Temperature is extremely below {temp} deg")

def Temperature(temp):
    if temp < 0:
        raise TooColdError(temp)
    elif temp >= 50:
        raise TooHotError(temp)
    else:
        return f"Temperature is moderate {temp} deg"

try:
    temperature = float(input("enter temperature: "))
    result = Temperature(temperature)
    print(result)
except TooHotError as e:
    print(e)
except TooColdError as e:
    print(e)
except ValueError:
    print(ValueError)
finally:
    print("program finished")
```
