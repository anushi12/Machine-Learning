### Abstraction

Abstraction is the concept of hiding the complex implementation details and showing only the necessary features of an object. This helps in reducing programming complexity and effort.

```python
## Abstract Base Class

from abc import ABC, abstractmethod

## this the template that every class must follow
class Vehicle(ABC):
    def drive(self):
        print("The vehicle is used for daily purpose.")
    
    @abstractmethod         ## this is an abstract method that is for implememting based on rules
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("car engine started")

def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()

car = Car()
operate_vehicle(car)
```

```python
class Bank(ABC):
    @abstractmethod
    def deposit(self,amount):
        pass

    @abstractmethod
    def withdraw(self,amount):
        pass


class SavingsAccount(Bank):
    def __init__(self, name, balance = 0):
        self.name = name
        self.balance = balance
    
    def deposit(self,amount):
        self.balance += amount
        print(f"Deposited: {amount}, New balance: {self.balance}")

    def withdraw(self,amount):
        if amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance -= amount
            print(f"withdrew: {amount}, New balance: {self.balance}")

person = SavingsAccount("anusha", 1000)
print(person.name, person.balance)
person.deposit(500)
person.withdraw(100)
```
