## Classes and Objects

Object-Oriented Programming (OOP) is a programming paradigm that uses "objects" to design applications and computer programs. OOP allows for modeling real-world scenarios using classes and objects. 

1. Class and Object

Class -> Blueprint
A class defines the structure and behavior (attributes + methods).

Object -> Instance of a class
An object is created from a class.

```python
class Car:
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color
    
    def start(self):
        print(f"{self.brand} in {self.color} is starting...")

my_car = Car("Porshe","Red")
my_car.start()
```

```python
## Instance variable and Method
class Dog:
    ## constructor
    def __init__(self,name,age):
        self.name = name
        self.age = age
    
## create objects
dog1 = Dog("Tyson",'2')
print(dog1.name)
print(dog1.age)
```

```python
## instance method

class Dog:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    ## method
    def bark(self):
        print(f"{self.name} is barking woof woof...")

dog1 = Dog("Tyson",2)
print(dog1.name, dog1.age)
dog1.bark()
```

```python
## Modeling a Bank Account

## define a class for bank account
class BankAccount:
    def __init__(self,name,age,balance=0):
        self.name = name
        self.age = age
        self.balance = balance
    def show_balance(self):
        print(f"Balance is {self.balance}")

    def deposit(self,amount):
        self.balance += amount
        print(f"{amount} is deposited and the current balance is {self.balance}")

    def withdraw(self,amount):
        try:
            if self.balance < amount:
                print("insufficient balance")
        except Exception as ex:
            print(ex)
        else:
            self.balance -= amount
            print(f"{amount} is withdrawed and the current balance is {self.balance}")

person1 = BankAccount("Puneet",65,2000)
print(person1.name, person1.age)
person1.show_balance()
person1.deposit(2000)
person1.withdraw(500)
person1.show_balance()

   
```
