# Sets

Sets are built-in data type in Pyhton used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference and symmetric difference.

```python
my_set = {1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set = set()
print(type(my_empty_set))
```

```python
my_set = set([1,3,5,5,6,6,6,3,8,9])
print(my_set)
```

```python
## basic set operation
## adding and removing elements

my_set.add(2)
print(my_set)
```

```python
## remove element from set

my_set.remove(6)
print(my_set)
```

```python
my_set.remove(11)       ## remove will throw an error if the element is not present in the set
print(my_set)
```

```python
my_set.discard(11)  ## discard will ignore if the element is not present
print(my_set)
```

```python
## pop      ---> removes the first element of the list (FIFO)

removed_element = my_set.pop()
print(removed_element)
print(my_set)
```

```python
## clear all the elements

my_set.clear()
print(my_set)
```

```python
## set membership test

my_set = {1,2,3,4,5,6}
print( 3 in my_set)
print( 11 in my_set)
```

```python
## Mathematical Operation

set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

## Union --> get all elements

union_set = set1.union(set2)
print(union_set)
```

```python
## intersection

intersection_set = set1.intersection(set2)
print(intersection_set)


## intersection_update ---> it will update the set too

set1.intersection_update(set2)
print(set1)
```

```python
## difference --> elements present in set 1 but not in set 2

set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}
print(set1.difference(set2))
```

```python
## symmetric difference --> unique elements from both the sets

print(set1.symmetric_difference(set2))
```

```python
## set method 
# issubset --> check if all elememt of set 1 is present in set 2
# issuperset --> checks if set1 contains all elements of set2

set1 = {1,2,3,4,5}
set2 = {3,4,5}

print(set1.issubset(set2))
print(set1.issuperset(set2))
```

```python
lst = [1,2,3,4,4,5,2]
set(lst)
```

```python
## counting unique words in text

text = " hey i am happy and enjoying learning pyhton"
words = text.split()

# converts list of words to set to get unique words

unique_words = set(words)
print(unique_words)
print(len(unique_words))
```
