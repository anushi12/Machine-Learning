Example 1. Manage a todo list

- Create a to do list to keep track of tasks

```python
to_do_lists = ["Clean house", "study", "pay bills"]

## Adding to task
to_do_lists.append("Schedule meeting")
to_do_lists.append("water the plants")

## removing a completed task
to_do_lists.remove("Clean house")

## checking if a task is in the list

if "pay bills" in to_do_lists:
    print("Don't forget to pay bills")

print("To do list remaining")

for task in to_do_lists:
    print(f'- {task}')
```

Example 2. Organising student grade


```python
grades = [75, 87, 99, 32, 56, 79]

# adding grades

grades.append(66)

# calculating the average grade

avg_grade = sum(grades)/len(grades)
print(f"Average grade: {avg_grade: .2f}")

## finding highest and lowest grades

higest_grade = max(grades)
lowest_grade = min(grades)

print(f"Highest grade : {higest_grade} and lowest grade : {lowest_grade}")
```

Example 3. Managing an inventory

- Use a list to manage inventory items in a store

```python
inventory = ["apples", "banana", "kiwi", "grapes"]

## adding new item
inventory.append("mango")

# removing an item that is out of stock
inventory.remove("apples")

# checking if an item is out of stock

item = "orange"
if item in inventory:
    print(f"{item} are in stock")
print(f"{item} is out of stock.")

print("inventory list:")
for item in inventory:
    print(f"- {item}")
```

Example 4. Customer Feedback Count

```python
feedback = ["Great service!", "Very satisfied", "could be better", "excellent experience"]

# adding a feedback
feedback.append("Not satisfied")

# counting specific feedback

positive_feedback = sum(1  for comment in feedback if "great" in comment.lower() or "excellent" in comment.lower())
print(f"Positive Feedback Count: {positive_feedback}")

## print all feedback

print("User feedback: ")
for comment in feedback:
    print(f" - {comment}")
```
