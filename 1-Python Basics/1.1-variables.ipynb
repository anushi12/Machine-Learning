## Variables

Variables are fundamental elements in programming used to store data that can be referenced and manipulated in a program.In python, variables are created when you assign a value to them, and they do not need explicit declaration to reserve meory space. The declaration happens automatically when you assign a value to a variable.

```python
## Declaring and Assining Variables

age = 32
height = 6
name = "Anusha"
is_student = True

## printing the variables

print("age : ", age)
print("height : ",height)
print("name : ",name)
print("is_student : ",is_student)
```

## Naming Conventions

- Variable names should be descriptive
- They must start with a letter or '_' and contains letters, numbers or underscores
- Variables in pyhton are case sensitive
- Variable names cannot be reserve keywords

```python
## valid variable names

first_name = "anusha"
_age = 32
name1 = "anusha"
```

```python
## invalid variable names

'''
2age = 30
@age = 30
first-name = "Anusha"
'''
```

```python
## Variable Types
## Python is dynamically typed language, type of a variable is determined at runtime.
age = 35
name = "Anusha"

print(type(age))
print(type(name))
```

```python
## input

age = input("what is your age?")
print(age, type(age))
age_int = int(input("enter age"))
print(age_int,type(age_int))
```

```python
### simple calculator

num1 = float(input("enter first number: "))
num2 = float(input("enter second number: "))

sum = num1 + num2
difference = num1-num2
product = num1 * num2
quotient = num1 % num2

print("sum : ", sum)
print("difference : ", difference)
print("product : ", product)
print("quotient : ", quotient)
```
