```python
## Arithmetic Operations

a = 10
b = 5

add_res = a + b
sub_res = a - b
mult_res = a * b
div_res = a / b
floor_res = a // b
mod_res = a % b
exponent_res = a**b

print(add_res)
print(sub_res)
print(mult_res)
print(div_res)
print(floor_res)
print(mod_res)
print(exponent_res)
```

## Comparison Operators

```python
## Comparison Operators

## == equal to operator
a = 10
b = 10

a == b
```

```python
str1 = "puneet"
str2 = "Puneet"

## it is case sensitive 
str1 == str2
```

```python
## != (Not equal to)
str1 != str2
```

```python
## greater than >

num1 = 45
num2 = 55

num1 > num2
```

```python
# smaller than <
num1 < num2
```

```python
## greater than or equal to

num1 = 90
num2 = 69

num1 >= num2
```

```python
## smaller than or equal to

num1 = 90
num2 = 90

num1 <= num2
```

## Logical Operators

```python
## AND, NOT, OR

x = True
y = True

x and y  ## In case of AND both cases needs to be true
```

```python
X = True
Y = False

X or Y ## in case of OR any one case needs to be true
```
